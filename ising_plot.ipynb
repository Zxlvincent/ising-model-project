```python
import pandas as pd
import matplotlib.pyplot as plt

# upload simulation data from CSV 
df1d = pd.read_csv("ising_1d.csv")
df2d = pd.read_csv("ising_2d.csv")
betas = sorted(df1d["Beta"].unique())  # Unique beta values used in simulation

#  Plot average value with error bars (mean  std) 
def plot_with_error(df, variable, title, color):
    """
    Plot average with error bars for a given variable (Energy or Magnetization)
    """
    stats = df.groupby("Beta")[variable].agg(['mean', 'std']).reset_index()
    plt.figure()
    plt.errorbar(stats["Beta"], stats["mean"], yerr=stats["std"], fmt='-o', color=color)
    plt.title(title)
    plt.xlabel("Beta ()")
    plt.ylabel(f"Average {variable}")
    plt.grid(True)
    plt.savefig(title.lower().replace(" ", "_") + ".png")

#  Mean & Error plots 
plot_with_error(df1d, "Energy", "1D: Average Energy vs Beta", "blue")
plot_with_error(df1d, "Magnetization", "1D: Average Magnetization vs Beta", "green")
plot_with_error(df2d, "Energy", "2D: Average Energy vs Beta", "red")
plot_with_error(df2d, "Magnetization", "2D: Average Magnetization vs Beta", "purple")

# Plot histograms (distribution of values at each beta) 
for beta in betas:
    for model, df in zip(["1d", "2d"], [df1d, df2d]):
        subset = df[df["Beta"] == beta]

        # Energy histogram
        plt.figure()
        plt.hist(subset["Energy"], bins=30)
        plt.title(f"{model.upper()}: Energy Distribution at Beta = {beta}")
        plt.xlabel("Energy")
        plt.ylabel("Frequency")
        plt.grid(True)
        plt.savefig(f"{model}_energy_dist_beta{beta}.png")

        # Magnetization histogram
        plt.figure()
        plt.hist(subset["Magnetization"], bins=30)
        plt.title(f"{model.upper()}: Magnetization Distribution at Beta = {beta}")
        plt.xlabel("Magnetization")
        plt.ylabel("Frequency")
        plt.grid(True)
        plt.savefig(f"{model}_magnet_dist_beta{beta}.png")

# Plot convergence graphs (Energy & Magnetization vs Step) 
for model, df in zip(["1d", "2d"], [df1d, df2d]):
    for beta_val in betas:
        subset = df[df["Beta"] == beta_val]

        # Energy vs steps
        plt.figure()
        plt.plot(subset["Step"], subset["Energy"])
        plt.title(f"{model.upper()}: Energy vs Step (Beta = {beta_val})")
        plt.xlabel("Step")
        plt.ylabel("Energy")
        plt.grid(True)
        plt.savefig(f"{model}_energy_steps_beta{beta_val}.png")

        # Magnetization vs steps
        plt.figure()
        plt.plot(subset["Step"], subset["Magnetization"])
        plt.title(f"{model.upper()}: Magnetization vs Step (Beta = {beta_val})")
        plt.xlabel("Step")
        plt.ylabel("Magnetization")
        plt.grid(True)
        plt.savefig(f"{model}_magnet_steps_beta{beta_val}.png")

print(" All plots generated successfully.")
```
